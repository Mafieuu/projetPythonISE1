# Etape 10
* Filtrer les équipes qui ont marqué plus de 6 buts

```python
# l'astuce viens des masques booleen. Les Series de pandas fonctionnent un peu comme les tableau numpy !
```

```python
team_goals_plus6=((euro12.loc[euro12.loc[:,"Goals"]>=6]).loc[:,"Team"]).values
# l'astuce viens des masques booleen
# on filtre les equipes qui ont marque plus de 6 but puis on regarde les valeurs de la colonne Team
```

```python
print(f" les equipes ayant inscript au moins 6 buts sont \n {team_goals_plus6}")
```

# Etape 11
* Sélectionnez les équipes commençant par G

```python
# on va detailler le code
```

```python
# On cree notre masque booleen
masque_name_G = euro12.loc[:,"Team"].apply(lambda x: x[0]) == 'G'
```

```python
# on affiche le resultat
# methode 1 ==> team_name_G=(euro12.loc[masque_name_G]).loc[:,"Team"]
team_name_G=euro12.loc[masque_name_G,"Team"]
```

```python
print(f"Les equipes commencant par G sont : \t {team_name_G.values} ")
```
